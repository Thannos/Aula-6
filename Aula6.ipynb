```python
import matplotlib.pyplot as plt
import numpy as np
def funcao(t):
    return 3-2*t 
def funcao1(t):
    return (-2-t)/-3
n = np.linspace(-2,12, num=1000)
plt.plot(n, funcao(n))
plt.plot(n, funcao1(n))
plt.show()
```

```python
a=np.array([[2,1],[1,-3]])
b=np.array([3,-2])
x=np.linalg.solve(a,b)
print(x)
#array([2., 3.])
np.allclose(np.dot(a,x),b)
```
